#### Usuwamy kopie zduplikowanych zdj ze zbioru

```python
hashes = set()
duplicates = 0

for category in categories:
    category_path = os.path.join(data_path, category)
    
    for img_name in os.listdir(category_path):
        img_path = os.path.join(category_path, img_name)
        
        with open(img_path, "rb") as f:
            img_hash = hashlib.md5(f.read()).hexdigest()
        
        if img_hash in hashes:
            duplicates += 1
            os.remove(img_path)  
            print(f"Usunito duplikat: {img_path}")
        else:
            hashes.add(img_hash)

print(f"Liczba usunitych duplikat贸w: {duplicates}")
```

#### Sprawdzamy licznoci obraz贸w w podziale wedug rozmiaru

#### Sprawdzamy czy obrazy nie s uszkodzone

```python
def check_missing_images(data_path, categories):
    missing_images = []
    for category in categories:
        category_path = os.path.join(data_path, category)
        for img_name in os.listdir(category_path):
            img_path = os.path.join(category_path, img_name)
            try:
                with Image.open(img_path) as img:
                    img.verify()  
            except (IOError, SyntaxError) as e:
                missing_images.append(img_path)
    
    return missing_images

missing_images = check_missing_images(data_path, categories)
print(f"Znaleziono {len(missing_images)} uszkodzonych obraz贸w.")
```

#### Normalizujemy obrazy do jednego rozmiaru, zostawiamy czarne pasy - bdziemy ich obecno uwzgldnia przy ekstarkcji cech

```python
def normalize_image(img_path):
    with Image.open(img_path) as img:
        img = img.resize((70, 100))  
        img_array = np.array(img) / 255.0 
        return img_array

sizes = []
for category in categories:
    category_path = os.path.join(data_path, category)
    for img_name in os.listdir(category_path):  
        img_path = os.path.join(category_path, img_name)
        img_array = normalize_image(img_path)  
        sizes.append(img_array.shape[:2])  
print(set(sizes))

normalized_images = []
for category in categories:
        category_path = os.path.join(data_path, category)
        img_name = random.choice(os.listdir(category_path))
        img_path = os.path.join(category_path, img_name)
        normalized_images.append(normalize_image(img_path))



def display_normalized_image(normalized_image):
       plt.imshow(normalized_image)
       plt.axis('off') 
       plt.show()
display_normalized_image(normalized_images[0])
```

```python
def image_brightness(img_path):
    with Image.open(img_path) as img:
        img_gray = img.convert("L")  
        return np.array(img_gray).mean()

brightness_values = []

for category in categories:
    category_path = os.path.join(data_path, category)
    img_name = random.choice(os.listdir(category_path))
    img_path = os.path.join(category_path, img_name)
    brightness_values.append(image_brightness(img_path))

plt.figure(figsize=(10, 5))
plt.hist(brightness_values, bins=20, color="orange", alpha=0.7)
plt.xlabel("Jasno")
plt.ylabel("Liczba obraz贸w")
plt.title("Histogram jasnoci obraz贸w")
plt.show()
```

##### Powtarzamy wczeniejsze kroki dla zmodyfikowanej funkcji liczcej redni kolor贸w

```python
def avg_color_without_white_and_black(img_path):
    """Oblicza redni warto kolor贸w RGB dla obrazu, ignorujc piksele poni偶ej (3,3,3) i powy偶ej (250,250,250)."""
    with Image.open(img_path) as img:
        img = img.convert("RGB")  
        img = img.resize((100, 70))  
        img_array = np.array(img)

        
        mask = np.all(img_array >= [3, 3, 3], axis=2) & np.all(img_array <= [250, 250, 250], axis=2)
        valid_pixels = img_array[mask]

        if len(valid_pixels) == 0:  
            return np.array([0, 0, 0])  

        avg_rgb = np.mean(valid_pixels.reshape(-1, 3), axis=0)  
        return avg_rgb
```

```python
df[["R2", "G2", "B2"]] = df["path"].apply(lambda x: pd.Series(avg_color_without_white_and_black(x)))
```

#### Badamy valid_pixels_ratio - stosunek kolorowych pikseli do caosi (miara wielkoci fasolki)

```python
def valid_pixels_ratio(img_path):
    """Oblicza stosunek pikseli speniajcych filtr (3,3,3) - (250,250,250) do cakowitej liczby pikseli."""
    with Image.open(img_path) as img:
        img = img.convert("RGB")  
        img_array = np.array(img)  
        
        total_pixels = img_array.shape[0] * img_array.shape[1] 
       
        mask = np.all(img_array >= [3, 3, 3], axis=2) & np.all(img_array <= [250, 250, 250], axis=2)
        valid_pixel_count = np.count_nonzero(mask)  

        return valid_pixel_count / total_pixels  
df["valid_pixel_ratio"] = df["path"].apply(valid_pixels_ratio)
```

```python
def image_contrast(img_path):
    """Oblicza kontrast obrazu jako odchylenie standardowe wartoci jasnoci pikseli."""
    with Image.open(img_path) as img:
        img = img.convert("L")  
        img_array = np.array(img)
        return np.std(img_array)  

for class_folder in os.listdir(data_path): 
    class_path = os.path.join(data_path, class_folder)
    if os.path.isdir(class_path):  
        for file in os.listdir(class_path):  
            if file.lower().endswith((".jpg", ".jpeg", ".png")):
                img_path = os.path.join(class_path, file)
                data.append({
                    "class": class_folder,
                    "filename": file,
                    "path": img_path,
                    "contrast": image_contrast(img_path)
                })


contrast_df = pd.DataFrame(data)


plt.figure(figsize=(10, 5))
ax = sns.histplot(contrast_df, x="contrast", hue="class", bins=30, kde=True, palette=colors, alpha=0.7)


handles, labels = ax.get_legend_handles_labels()
ax.legend_.remove()  

print("Legend labels:", labels)


if not labels:
    labels = contrast_df["class"].unique().tolist()
    handles = [plt.Line2D([0], [0], color=sns.color_palette("tab20")[i], lw=4) for i in range(len(labels))]


plt.legend(handles, labels, title="Klasy", loc="center left", bbox_to_anchor=(1, 0.5))


plt.xlabel("Kontrast")
plt.ylabel("Liczba obraz贸w")
plt.title("Histogram kontrastu obraz贸w")


plt.tight_layout(rect=[0, 0, 0.85, 1])


plt.show()
```

##### Dla wikszoci kategorii - rozkady zbli偶one do normalnych o r贸偶nych rednich
##### zastanawia wzrost liczby obserwownaych przypadk贸w dla najwikszej wartoci 120

#### Poniewa偶 w powy偶szym przypadku na wynik r贸wnie偶 wpyway czarne pasy oraz stosunek obszaru zajmowanego przez fasolk do caoci zdjcia, zbadajmy kontrast bez tych element贸w  (filtr jak wczeniej)

```python
def image_contrast_without_extreme_values(img_path):
    """Oblicza kontrast obrazu jako odchylenie standardowe wartoci jasnoci pikseli,
    ignorujc piksele o wartociach <3 i >250.
    """
    with Image.open(img_path) as img:
        img = img.convert("L")  # Konwersja do skali szaroci
        img_array = np.array(img)

        # Usuwamy piksele o wartociach <3 i >250
        valid_pixels = img_array[(img_array >= 3) & (img_array <= 250)]

        if valid_pixels.size > 0:
            return np.std(valid_pixels)  # Kontrast = odchylenie standardowe wartoci jasnoci
        else:
            return 0  # Jeli nie ma 偶adnych wa偶nych pikseli, zwracamy 0


data = []


for class_folder in os.listdir(data_path): 
    class_path = os.path.join(data_path, class_folder)
    if os.path.isdir(class_path):  
        for file in os.listdir(class_path):  
            if file.lower().endswith((".jpg", ".jpeg", ".png")):
                img_path = os.path.join(class_path, file)
                data.append({
                    "class": class_folder,
                    "filename": file,
                    "path": img_path,
                    "contrast2": image_contrast_without_extreme_values(img_path)
                })
```
