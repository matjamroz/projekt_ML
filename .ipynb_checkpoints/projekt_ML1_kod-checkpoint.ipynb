```python
import matplotlib.pyplot as plt
from PIL import Image
import random
```

```python
def plot_sample_images(data_path, categories, n=5):
    fig, axes = plt.subplots(1, n, figsize=(15, 5))
    for i in range(n):
        category = random.choice(categories)  # Wybieramy losową kategorię
        img_name = random.choice(os.listdir(os.path.join(data_path, category)))  # Losowy obraz
        img_path = os.path.join(data_path, category, img_name)
        img = Image.open(img_path)
        axes[i].imshow(img)
        axes[i].set_title(category)
        axes[i].axis("off")
    plt.show()

plot_sample_images(data_path, categories)
```

```python
sizes = []
for category in categories:
    category_path = os.path.join(data_path, category)
    for img_name in os.listdir(category_path):  
        img_path = os.path.join(category_path, img_name)
        with Image.open(img_path) as img:
            sizes.append(img.size)

print(set(sizes)) 
```

```python
from collections import defaultdict

# Tworzymy słownik, który będzie liczył wystąpienia rozmiarów
size_count = defaultdict(int)

# Iterujemy po obrazach
for category in categories:
    category_path = os.path.join(data_path, category)
    for img_name in os.listdir(category_path):
        img_path = os.path.join(category_path, img_name)
        with Image.open(img_path) as img:
            size = img.size
            size_count[size] += 1  # Zwiększamy licznik dla danego rozmiaru

# Wyświetlamy liczbę obrazów dla każdego rozmiaru
for size, count in size_count.items():
    print(f"Rozmiar {size}: {count} obrazów")
```

```python
from PIL import Image

def check_missing_images(data_path, categories):
    missing_images = []
    for category in categories:
        category_path = os.path.join(data_path, category)
        for img_name in os.listdir(category_path):
            img_path = os.path.join(category_path, img_name)
            try:
                with Image.open(img_path) as img:
                    img.verify()  # Sprawdzenie, czy obraz jest poprawny
            except (IOError, SyntaxError) as e:
                missing_images.append(img_path)
    
    return missing_images

missing_images = check_missing_images(data_path, categories)
print(f"Znaleziono {len(missing_images)} uszkodzonych obrazów.")
```

```python
def image_brightness(img_path):
    with Image.open(img_path) as img:
        img_gray = img.convert("L")  # Konwersja na grayscale
        return np.array(img_gray).mean()  # Średnia jasność

brightness_values = []

for category in categories:
    category_path = os.path.join(data_path, category)
    img_name = random.choice(os.listdir(category_path))
    img_path = os.path.join(category_path, img_name)
    brightness_values.append(image_brightness(img_path))

plt.figure(figsize=(10, 5))
plt.hist(brightness_values, bins=20, color="orange", alpha=0.7)
plt.xlabel("Jasność")
plt.ylabel("Liczba obrazów")
plt.title("Histogram jasności obrazów")
plt.show()
```

```python
import hashlib

hashes = set()
duplicates = 0

for category in categories:
    category_path = os.path.join(data_path, category)
    
    for img_name in os.listdir(category_path):
        img_path = os.path.join(category_path, img_name)
        
        with open(img_path, "rb") as f:
            img_hash = hashlib.md5(f.read()).hexdigest()
        
        if img_hash in hashes:
            duplicates += 1
        else:
            hashes.add(img_hash)

print(f"Liczba zduplikowanych obrazów: {duplicates}")
```

```python
import numpy as np
import os
import random
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

def avg_color(img_path):
    """Oblicza średnią wartość kolorów RGB dla pojedynczego obrazu."""
    with Image.open(img_path) as img:
        img = img.resize((50, 50))  # Zmniejszenie dla optymalizacji
        img_array = np.array(img)
        return np.mean(img_array, axis=(0, 1))  # Średnia dla R, G, B

# Liczba obrazów do analizy na kategorię
NUM_IMAGES = 80  
```

```python
colors = []

for category in categories:
    category_path = os.path.join(data_path, category)
    image_files = os.listdir(category_path)
    
   
    selected_images = random.sample(image_files, min(NUM_IMAGES, len(image_files)))
    
    category_colors = [avg_color(os.path.join(category_path, img)) for img in selected_images]
    
    
    avg_category_color = np.mean(category_colors, axis=0)
    
    colors.append(avg_category_color)

df_colors = pd.DataFrame(colors, columns=["Red", "Green", "Blue"], index=categories)
df_colors.plot(kind="bar", figsize=(12, 6), colormap="viridis")
plt.title("Średnie wartości RGB dla każdej kategorii (na podstawie wielu obrazów)")
plt.show()
```

```python
def normalize_image(img_path):
    with Image.open(img_path) as img:
        img = img.resize((50, 50))  # Opcjonalne zmniejszenie obrazu
        img_array = np.array(img) / 255.0  # Normalizacja do [0, 1]
        return img_array

# Normalizowanie obrazów
normalized_images = []
for category in categories:
    category_path = os.path.join(data_path, category)
    img_name = random.choice(os.listdir(category_path))
    img_path = os.path.join(category_path, img_name)
    normalized_images.append(normalize_image(img_path))

# Możesz teraz zapisać te obrazy w formie macierzy, np. jako dane wejściowe do modelu
```

```python
def display_normalized_image(normalized_image):
    # Znormalizowany obraz jest w postaci (50, 50, 3), musimy go przywrócić do formatu, który będzie można wyświetlić
    plt.imshow(normalized_image)
    plt.axis('off')  # Wyłączamy osie
    plt.show()

# Wyświetlamy pierwszy znormalizowany obraz
display_normalized_image(normalized_images[0])
```
