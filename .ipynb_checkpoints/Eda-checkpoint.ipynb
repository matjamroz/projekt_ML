```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import random
import seaborn as sns
```

```python
data_path="Dataset"

categories=os.listdir(data_path)
```

```python
for category in categories:
    category_path = os.path.join(data_path, category)
    if os.path.isdir(category_path):
        print(f"{category}: {len(os.listdir(category_path))} images")
```

```python
# Pobranie listy plików i folderów
data = []
for class_folder in os.listdir(data_path):  # Iteracja po folderach klas
    class_path = os.path.join(data_path, class_folder)
    if os.path.isdir(class_path):  # Sprawdzenie, czy to folder
        for file in os.listdir(class_path):  # Iteracja po plikach w folderze
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):  # Filtrujemy obrazy
                data.append({"class": class_folder, "filename": file, "path": os.path.join(class_path, file)})

# Tworzenie DataFrame
df = pd.DataFrame(data)

# Podgląd danych
print(df.head())
```

```python
def image_brightness(img_path):
    with Image.open(img_path) as img:
        img_gray = img.convert("L")  # Konwersja na grayscale
        return np.array(img_gray).mean()  # Średnia jasność
```

```python
def avg_color(img_path):
    """Oblicza średnią wartość kolorów RGB dla pojedynczego obrazu."""
    with Image.open(img_path) as img:
        img = img.resize((50, 50))  # Zmniejszenie dla optymalizacji
        img_array = np.array(img)
        return np.mean(img_array, axis=(0, 1))  # Średnia dla R, G, B
```

```python
def normalize_image(img_path):
    with Image.open(img_path) as img:
        img = img.resize((50, 50))  # Opcjonalne zmniejszenie obrazu
        img_array = np.array(img) / 255.0  # Normalizacja do [0, 1]
        return img_array
```

```python
def display_normalized_image(normalized_image):
    # Znormalizowany obraz jest w postaci (50, 50, 3), musimy go przywrócić do formatu, który będzie można wyświetlić
    plt.imshow(normalized_image)
    plt.axis('off')  # Wyłączamy osie
    plt.show()
```

```python
df["brightness"] = df["path"].apply(image_brightness)
```

```python
df[["R", "G", "B"]] = df["path"].apply(lambda x: pd.Series(avg_color(x)))
```

```python
# Ustawienie stylu wykresów
sns.set_style("whitegrid")

# Tworzymy wykresy dla każdej cechy
features = ["brightness", "R", "G", "B"]

plt.figure(figsize=(12, 8))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=df, x=feature, hue="class", kde=True, bins=30, palette="tab20", alpha=0.6)
    plt.title(f"Rozkład {feature} dla każdej klasy")
    plt.xlabel(feature)
    plt.ylabel("Liczność")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, x="class", y=feature, palette="coolwarm")
    plt.xticks(rotation=45)
    plt.title(f"Rozkład {feature} w różnych klasach")

plt.tight_layout()
plt.show()
```
